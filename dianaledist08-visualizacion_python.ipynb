### Ejecución del archivo .csv

### Variables de la base de datos

*   gender: Género - 0 (desconocido), 1 (varón), 2 (mujer)
*   C_api:  género extraído de la API de WikiMedia, codificado como femenino / masculino / desconocido
*   C_man: género extraído de la codificación de contenido, codificado como 1 (varón) / 2 (mujer) / 3 (desconocido)
* E_NEds: Índice I del estrato IJ (0, 1, 2, 3)
* E_Bpag: Índice J del estrato IJ (0, 1, 2, 3)
* firstDay: primera edición en la Wikipedia en español (AAAA/MM/DD/HH/MM/SS)
* lastDay: última edición en la Wikipedia en español (AAAA/MM/DD/HH/MM/SS)
* NEds: número total de ediciones
* NDays: número de días (lastDay - firstDay + 1)
* NActDays: número de días con ediciones
* NPages: número de páginas diferentes editadas
* NPcreated: número de páginas creadas
* pagesWomen: número de ediciones en páginas relacionadas con mujeres
* wikiprojWomen: número de ediciones en WikiProjects relacionados con mujeres
* ns_user: número de ediciones en el espacio de nombres de las personas usuarias
* ns_wikipedia: número de ediciones en el espacio de nombres de Wikipedia
* ns_talk: número de ediciones en el espacio de nombres de discusión
* ns_userTalk: número de ediciones en el espacio de nombres de discusión de las personas usuarias
* ns_content: número de ediciones en páginas de contenido
* weightIJ: peso de corrección para el estrato IJ
* NIJ: número de elementos en el estrato IJ

## Visualización de gráficos


## matplotlib

### Histograma

```python
plt.figure(figsize=(10,6))
plt.hist(df['gender'], bins=5, color='purple', edgecolor='black')
plt.title('¿Cómo se distribuyen las personas editoras en Wikipedia según su género?')
plt.xlabel('Género de las personas editoras y autoras')
plt.ylabel('Personas editoras y autoras en WP')
plt.xticks([0, 1, 2], ["0=Género no declarado", "1=Varón", "2=Mujer"])
plt.show()
```

### Explicación:

El gráfico muestra una visión general del género de las personas que escriben o editan entradas en Wikipedia. Para las cuentas con género identificado, la mayoría refieren a varones y hay una baja aparición de perfiles identificados como mujeres.


## matplotlib

### Scatterplot

```python
plt.figure(figsize=(15, 8))
mapeo_colores = plt.scatter(df['NPages'], df['pagesWomen'], c=df["gender"], cmap='viridis', alpha=0.5)
plt.colorbar(mapeo_colores)
plt.title('Ediciones en Wikipedia vs. Ediciones de entradas referidas a mujeres')
plt.xlabel('Ediciones hechas por persona')
plt.ylabel('Ediciones en páginas referidas a mujeres')
plt.show()
```

### Explicación:

El gráfico muestra una representación gráfica de la cantidad de ediciones que se realizan en Wikipedia por cada persona en función de las ediciones que refieren a entradas sobre mujeres. En primer lugar, se observa que hay una baja proporción de entradas que refieren a mujeres. Y a pesar de la baja existencia de perfiles identificados como mujeres (puntos amarillos), realizan más ediciones en entradas sobre mujeres.


## matplotlib

### Lineplot

```python
plt.figure(figsize=(12, 6))
plt.plot(df['NEds'], label='Total de ediciones hechos por persona', color='pink', marker='o', linestyle='-', linewidth=1)
plt.title('Lineplot de ediciones hechas por persona')
plt.xlabel('Índice de autoría')
plt.ylabel('Cantidad de ediciones')
plt.legend()
plt.show()
```

### Explicación:

El gráfico muestra la cantidad de ediciones realizadas por persona. Se observa que la mayor cantidad de perfiles en Wikipedia han realizado menos de 20000 ediciones en la plataforma.


## seaborn

### Histograma

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='C_man', data=df, palette='viridis')
plt.title('Cantidad de personas editoras por Género extraído del código de contenido')
plt.xlabel('Género extraído del código')
plt.ylabel('Cantidad de personas editoras')
plt.xticks([0, 1, 2], ["1=Varón", "2=Mujer", "3=Género no identificado"])
plt.show()
```

### Explicación:

El gráfico muestra la cantidad de ediciones realizadas por persona en función del género extraído del código de contenido. Similar a lo obtenido en el histograma realizado en Matplotlib, hay una baja aparición de mujeres editoras incluso en función de un género que se extrae manualmente del código de contenido.


## seaborn

### Barplot

```python
plt.figure(figsize=(20, 10))
plt.title('Cantidad de páginas editadas y páginas nuevas en función del Género')
sns.barplot(x="gender", y="NPcreated", hue="gender", data=df, palette='viridis')
sns.barplot(x="gender", y="NPages", hue="gender", data=df, palette='viridis', alpha=0.5)

plt.xlabel('Género')
plt.ylabel('Cantidad de nuevas páginas creadas (color plano) y Ediciones totales (color traslúcido)')

plt.xticks([0, 1, 2], ["0=Género no declarado", "1=Varón", "2=Mujer"])
```

### Explicación:

El gráfico muestra la cantidad de ediciones de entrada y de creación de nuevas páginas según el género declarado en el perfil de Wikipedia. En cuanto a las ediciones, tanto varones como mujeres editan entradas con similar frecuencia. Sin embargo, si tomamos en cuenta las páginas creadas,lo que es análogo a realizar el primer paso para que una entrada tenga existencia en la plataforma, las mujeres crean menos páginas que los varones por lo que su iniciativa es menor.


## seaborn

### Scatterplot

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x='lastDay', y='NDays', data=df, hue='gender', palette='viridis', alpha=0.7)
plt.title('Scatterplot del último día de edición en Wikipedia vs. Cantidad de días editando')
plt.xlabel('Último día de edición')
plt.ylabel('Cantidad de días editando')
plt.legend(title='Género')
plt.show()
```

### Explicación:

El gráfico muestra el último día editando en Wikipedia en función de la cantidad de días editando en la plataforma, según el género identificado de cada perfil. Las personas que más tiempo han estado editando en la plataforma son las que muestran su última edición en la plataforma más reciente en el tiempo.
