## Visualizaci칩n de gr치ficos


## matplotlib

### Histograma

```python
plt.figure(figsize=(10,6))
plt.hist(df['gender'], bins=5, color='purple', edgecolor='black')
plt.title('쮺칩mo se distribuyen las personas editoras en Wikipedia seg칰n su g칠nero?')
plt.xlabel('G칠nero de las personas editoras y autoras')
plt.ylabel('Personas editoras y autoras en WP')
plt.xticks([0, 1, 2], ["0=G칠nero no declarado", "1=Var칩n", "2=Mujer"])
plt.show()
```

### Explicaci칩n:

El gr치fico muestra una visi칩n general del g칠nero de las personas que escriben o editan entradas en Wikipedia. Para las cuentas con g칠nero identificado, la mayor칤a refieren a varones y hay una baja aparici칩n de perfiles identificados como mujeres.


## matplotlib

### Scatterplot

```python
plt.figure(figsize=(15, 8))
mapeo_colores = plt.scatter(df['NPages'], df['pagesWomen'], c=df["gender"], cmap='viridis', alpha=0.5)
plt.colorbar(mapeo_colores)
plt.title('Ediciones en Wikipedia vs. Ediciones de entradas referidas a mujeres')
plt.xlabel('Ediciones hechas por persona')
plt.ylabel('Ediciones en p치ginas referidas a mujeres')
plt.show()
```

### Explicaci칩n:

El gr치fico muestra una representaci칩n gr치fica de la cantidad de ediciones que se realizan en Wikipedia por cada persona en funci칩n de las ediciones que refieren a entradas sobre mujeres. En primer lugar, se observa que hay una baja proporci칩n de entradas que refieren a mujeres. Y a pesar de la baja existencia de perfiles identificados como mujeres (puntos amarillos), realizan m치s ediciones en entradas sobre mujeres.


## matplotlib
